```python
# Data quality checks
# if something returns a numpy.bool, you dont need to equate to anything
# use if condition: run the block, if false:    if not condition: 

# the validation table will allow us 




def data_check(df):
    """Checks for missing and duplicate values
    """
    
    if df.isnull().sum().sum():
        print("There is missing data in the df")
        # check the % of missing data
        print(f"% of missing values in each column {df.isnull().mean()}")
    else:
        print("No missing values")
    
    if df.duplicated().values.any():
        print("There are duplicate values")
    else:
        print("No duplicate values")
   
validation_rules = {
    'AGE': {'min': 18, 'max': 100, 'reason': 'Legal age for credit card'},
    'LIMIT_BAL': {'min': 0, 'max': 10000000, 'reason': 'Credit limits in NT dollars'},
    'EDUCATION': {'valid_values': [1, 2, 3, 4], 'reason': 'Documented categories'},
    'MARRIAGE': {'valid_values': [1, 2, 3], 'reason': 'Documented categories'},
    'PAY_0': {'min': -2, 'max': 9, 'reason': 'Payment status encoding'},
}     
    
        
def data_validation(df, validation_rules):
    """ - Checks for:
        - correct data types
        - logical range/boundaries of the features, 
    """
    report = []
    for col, criteria in validation_rules.items():
        if col not in df.columns:
            continue  # skip the column if not need to check
        
        # range check 
        if "min" in criteria and "max" in criteria:
            valid_mask = (df[col]>=criteria.get("min")) & (df[col]<=criteria.get("max"))
        
        # range check means continuos feature, so if that runs, that the categorical one is not applicable.
        elif "valid_values" in criteria:
            valid_mask = df[col].isin(criteria["valid_values"])
        else:
            continue
        # If we have a mask, identify the INVALID rows (where valid_mask is False)
        if valid_mask is not None:
            invalid_mask = ~valid_mask
            invalid_count = invalid_mask.sum()
            
            if invalid_count > 0:
                report.append({
                    'Column': col,
                    'Invalid_Count': invalid_count,
                    'Reason': criteria.get('reason', 'Value out of bounds'),
                    'Sample_Invalids': df.loc[invalid_mask, col].unique()[:5].tolist()
                })
    return report
    
    
```

```python
# df.isnull().sum().sum()

data_check(df)
data_validation(df,validation_rules)




```
